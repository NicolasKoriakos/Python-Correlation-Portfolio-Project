```python
# Import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
sns.set_style('white')

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) # Adjust the configuration of the plots

pd.set_option('display.max_rows', None)
```

```python
# Read the data

df = pd.read_csv(r'C:\Users\LENOVO\Downloads\movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# Check for missing data

for col in df.columns:
    porcent_missing = np.mean(df[col].isnull())
    print(f'{porcent_missing}% of {col} column is a missing value.')

```

```python
# Data types of our columns

df.dtypes
```

```python
# Changing columns datatype

# Creating a new df so we don't drop anything from the original dataset
df2 = df.dropna()

df2['budget']  = df2['budget'].astype('int64')
df2['gross']  = df2['gross'].astype('int64')
```

```python
df2.head()
```

```python
# Create correct year column

df2['CorrectYear'] = df2['released'].astype(str).str[:4]
```

```python
# Sort

df2 = df2.sort_values(by=['gross'], inplace=False, ascending=False)
```

Let's see if there is any correlation between budget and revenue 

```python
# Scatterplot Budget vs Gross revenue

plt.scatter(x=df2['budget'], y=df2['gross'])
plt.title('Budget vs Gross Revenue.')
plt.xlabel('Budget')
plt.ylabel('Gross Revenue')
plt.show()
```

```python
# Plot Budget vs Gross using seaborn

sns.regplot(x='budget', y='gross', data=df2, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Let's start looking at correlation

df2.corr(method='pearson')
```

```python
# High correlation between Budget and gross
# High correlation between votes and gross

correlation_matrix = df2.corr(method='pearson')
plt.title('Correlation matrix for numeric features.')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
# Let's check if there is correlation using the non-numerical values.

df_numerized = df2

for colname in df_numerized.columns:
    if df_numerized[colname].dtype == 'object':
        df_numerized[colname] = df_numerized[colname].astype('category')
        df_numerized[colname] = df_numerized[colname].cat.codes
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
plt.title('Correlation matrix for numeric features.')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_matrix_df_numerized = df_numerized.corr()
corr_pairs = correlation_matrix_df_numerized.unstack()
corr_pairs
```

```python
# Sort 
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Let's look for high correlations only
high_corr = sorted_pairs[sorted_pairs > 0.5]
high_corr
```
